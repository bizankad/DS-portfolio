# how to test if a time series is stationary?


Augmented Dicky-Fuller test
Null hypothesys is time series is non-stationary..


C'est tjrs bien de plotter car des exemple peuvent étre flaggé comme stationaire alors qu'il le sont pas (ex: lorsq'il n'y pas de trend mais tout de meme on a une variance qui change en fonction du temps)

Comment rendre une serie temporelle stationnaire?
- On fait la difference entre chaque observation et l'observation precedente (on perd le premier point)
- Pour certain serie temporelle on pourrais avoir besoin de faire la diffrence plus d'une fois

il se peut aussi qu'on ait besoin de faire np.log(df) ou np.sqrt(df) ou de calculer le changement proportionnelle = df.shift(1)/df -> **permet de recentrer les outliers autour de la valeur moyenne**

```python
amazon = pd.read_csv('amazon_close.csv', parse_dates=['date'], index_col=['date'])
```

```python
amazon
```

```python
# Calculate the first difference and drop the nans
amazon_diff = amazon.diff().dropna()
amazon_diff = amazon_diff.dropna()

# Run test and print
result_diff = adfuller(amazon_diff['close'])
print(result_diff)
```

```python
amazon_diff.plot()
```

On constante que le test de ducket n'est pas tres pertinent pour detecter le changement de la variance en fonction du temps. Cette serie n'est donc tjrs pas stationnaire.

```python
# Calculate the first difference and drop the nans
amazon_diff = amazon.diff()
amazon_diff = amazon_diff.dropna()

# Run test and print
result_diff = adfuller(amazon_diff['close'])
print(result_diff)

# Calculate log-return and drop nans
amazon_log = np.log((amazon/amazon.shift(1)).dropna())
amazon_log = amazon_log.dropna()

# Run test and print
result_log = adfuller(amazon_log['close'])
print(result_log)
```

```python
amazon_log.plot()
```

Great work! Notice that both the differenced and the log-return transformed time series have a small p-value, but the log transformed time series has a much more negative test statistic. This means the log-return tranformation is better.

# AR, MA and ARMA models

AR(p) (Autoregressive model) d'ordre 1(time lags: pas de temps): l'observation à l'intant t est égale à l'observation à l'instant t-1 multiplier par un coef + fluctuation (bruit blanc)
Si on fait l'analgie avec une regression linéaire le coef c'est ce que le modle apprend et la l'observation t-1 et une variable explicative.

AR(2) -> l'observation a l'instant t dependant de deux observation des deux pas de temps précedent + le bruit blanc. 



MA(q): l'observation à l'instant t est correlé avec l'error au instant précedents + bruit. Le modèle MA essay de compensé les erreur du passé.

ARMA = AR(p) + MA(q)

## SARIMAX

- Peut faire tout ce que ARMA peut faire.
- SARIMAX(df, order =(p,0,q), trend='c') == ARMA(df, order(p,q))
- trend='c' dois ajouter si la timeserie n'est pas centré autour de 0.


```python
earthquake.earthquakes_per_year.plot()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(earthquake.earthquakes_per_year, order=(2, 0, 1), trend='c')

results = model.fit()
# Generate predictions

# in-sample -> dans le train
one_step_forecast = results.get_prediction(start=-30)

# Extract prediction mean
mean_forecast = one_step_forecast.predicted_mean

# Get confidence intervals of predictions
confidence_intervals = one_step_forecast.conf_int()
```

```python
# Select lower and upper confidence limits
lower_limits = confidence_intervals.loc[:,'lower earthquakes_per_year']
upper_limits = confidence_intervals.loc[:,'upper earthquakes_per_year']

# Print best estimate  predictions
print(mean_forecast)
```

The predictions told me you could do it! You can use theis one-step-ahead forecast to estimate what your error would be, if you were to make a prediction for the Amazon stock price of tomorrow.

```python
# plot the amazon data
plt.plot(earthquake.earthquakes_per_year.index, earthquake.earthquakes_per_year.values, label='observed')

# plot your mean predictions
plt.plot(mean_forecast.index, mean_forecast.values, color='r', label='forecast')

# shade the area between your confidence limits
plt.fill_between(mean_forecast.index, lower_limits, upper_limits, color='pink')

# set labels, legends and show plot
plt.xlabel('Date')
plt.ylabel('Amazon Stock Price - Close USD')
plt.legend()
plt.show()
```

# ARIMA MODEL 

Le i de ARIMA est pour integrated et ca permet de de transformer une time series en ts stationaire. Ensuite après que la prediction soit faite sur la ts stationnaire, il fait une cumsum pour reconstruire la valeur de la prédiction dans la ts non-stationaire oringinal. 


d: ordre de differentiation

```python
amazon
```

```python
# Take the first difference of the data
amazon_diff = amazon.diff().dropna()

# Create ARMA(2,2) model
arma = SARIMAX(amazon_diff ,order=(2, 0, 2))

# Fit model
arma_results = arma.fit()

# Print fit summary
print(arma_results.summary())
```

Great. Remember that now you have a model which you can use to predict the difference values, not the absolute value of the amazon dataset.

```python
# Make arma forecast of next 10 differences
arma_diff_forecast = arma_results.get_forecast(steps=10).predicted_mean

# Integrate the difference forecast
arma_int_forecast = np.cumsum(arma_diff_forecast) 

# Make absolute value forecast
arma_value_forecast = arma_int_forecast + amazon.iloc[-1,0]

# Print forecast
print(arma_value_forecast)
```

Perfect. You have just made an ARIMA forecast the hard way. Next you'll use statsmodels to make things easier.

## ACF and PACF

ACF :


vocab : Offset d'un pas de temps

- offset lag-1 : correlation de la ts avec elle meme avec un offset d'un pas de temps.
- offset lag-2 : correlation de la ts avec elle meme avec un offset de deux pas de temps.


PACF : 

- Correlation associé uniquement avec ce lag
- c'est la correlation en ts et elle même offset de x pas de temps aprés qu'on ai soustrait l'effet des pas de temps précedent (les petit pas de temps)
Exemple correlation de ts lag-3: est la correlation entre ts et elle meme décalé de 3 pas de temps après qu'on soustrait l'effet de lag-2, et lag-1

# AIC and BIC

AIC:

- Lower AIC indicates a better model
- AIC likes to choose simple models with lower order (si on defini un trop élévé a notre modele par rapport au donnée l'AIC sera trop elevé -> cela peremet de controler l'overfitting)


BIC (Bayesian information criterion

- very similar to AIC 
- Lower BIC indicates a better model
- BIC likes to choose simple models with lower ordrer 


Difference entre BIC et AIC:

BIC favors simpler models than AIC (BIC penalise plus la complexité d'un modèle)


- AIC is better at choosing predictive models
- BIC is better at choosing good exlanatory model (car les modele seront plus simple donc plus interpretable)



On peut donc faire du grid search et utiliser AIC ou BIC pour trouver les parameters qui peremtte de les minimiser.

DAns le grid search certaine combin de p et q peuvent aboutir à une erreur -> car ne sont pas adapté au donnée donc tjrs utiliser un try exepth

**Searching over model order**

In this exercise you are faced with a dataset which appears to be an ARMA model. You can see the ACF and PACF in the plot below. In order to choose the best order for this model you are going to have to do a search over lots of potential model orders to find the best set.



```python
# Create empty list to store search results
order_aic_bic=[]

# Loop over p values from 0-2
for p in range(3):
  # Loop over q values from 0-2
    for q in range(3):
      	# create and fit ARMA(p,q) model
        model = SARIMAX(earthquake.earthquakes_per_year, order=(p, 0, q))
        results = model.fit()
        
        # Append order and results tuple
        order_aic_bic.append((p, q, results.aic, results.bic))
```

```python
# Construct DataFrame from order_aic_bic
order_df = pd.DataFrame(order_aic_bic, 
                        columns=['p', 'q', 'AIC', 'BIC'])

# Print order_df in order of increasing AIC
order_df.sort_values(by='AIC')
```

Fantastic! You built 9 models in just a few seconds! In the next exercise you will evaluate the results to choose the best model.



```python
# Loop over p values from 0-2
for p in range(3):
    # Loop over q values from 0-2
    for q in range(3):
      
        try:
            # create and fit ARMA(p,q) model
            model = SARIMAX(earthquake.earthquakes_per_year, order=(p, 0, q), freq=None)
            results = model.fit()
            
            # Print order and results
            print(p, q, results.aic, results.bic)
            
        except:
            print(p, q, None, None)     
```

Super! If you look at your printed results you will see that the AIC and BIC both actually favor an ARMA(1,1) model. This isn't what you predicted from the ACF and PACF but notice that the lag 2-3 PACF values are very close to significant, so the ACF/PACF are close to those of an ARMA(p,q) model.

# Model diagnostics
How good is the model we choose? 

Pour faire le diag de notre modèle, on s'intéresse au résidu. il s'agit de l'erreur entre les valeurs predit et les vrai valeurs.


- On peu acceder au résidus du model : resultas.resid
- on peu calculer la MAE de nos résidut: np.mean(np.abs(residuals))

Pour un modèle ideal les residu doit avoir la forme d'un bruit blanc gaussien centré autour de zero.

S'il existe un pattern dans le plot des residut, c'est qu'il s'agit d'info que le modèle n'a pas réussi à capturer.

```python
results.plot_diagnostics()
```

```python
results.summary()
```

- Prob(Q) : p-value for null hypo that residuals are uncorelated.  (test de Ljung-Box) 
- Prob(JB): p-value for null hypo that residuals are normal.(test de Jarque-Bera)

```python
# Fit model
model = SARIMAX(earthquake.earthquakes_per_year, order=(1,0,1))
results = model.fit()

# Calculate the mean absolute error from residuals
mae = np.mean(np.abs(results.resid))

# Print mean absolute error
print(mae)

# Make plot of time series for comparison
plt.figure(figsize=(20,6))
results.plot_diagnostics()
plt.show()
```

- If residuals are not normally distributed maybe we should increase d.
- if The residuals are correlated. You should increase p or q.

## SARIMA models 

Seasonal ARIMA = SARIMA


Si dans le residus on a des correlation, peut être qu'il existe une composante sasionniere. DAns ce cas utiliser Sarima pour la capturer.

SARIMA(p,d,q)(P,D,Q)

- (p,d,q) -> non seasonal orders for AR(p) and diff and MA(q)
- (P,Q,Q) ->  seasonal orders for seasonal autoregressive, seanonal differencing order, and seasonal moving av order + new order S: taille de la saison

pour la partie saisonniere le pas de temps est égal à la saison.

Si on construit un modèle SARIMA sans la partie non saisonnere, on capturera que la saisonalité mais les pattern locaux on ne les captruera pas.


seanonal differencing order:
- on fait la diff non pas entre le pas de temps t avec t-1, mais le pas de t avec le pas de temps t-s avec s etant la taille de la saison.
 -> df.diff(s)
 
 
Process pour determiner d, et D
- Si la ts montre une trend, on commence par un difference normal
- Si ensuite on contate le resultat montrer une seasonalité, on fait du seanonal differencing 


Process pour determiner P,Q, p, q
- ACF et PACF sauf que pour p et Q et specifier la liste des multiple de la saison dans le param lags
