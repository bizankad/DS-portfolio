# One-step prediction using  AutoRegressive Integrated Moving Average 

ARIMA = AR(p) + I(d) + MA(q)


- p = order of the autoregressive part;
- d = degree of first differencing involved;
- q = order of the moving average part.

In an autoregression model, we forecast the variable of interest using a linear combination of past values of the variable. The term autoregression indicates that it is a regression of the variable against itself.

The moving average part will use forecast errors in a regression like model. each predicted value can be thought of as a weighted moving average of the past few forecast errors.
the moving average corrects future forecasts based on errors made on recent forecasts.


```python
# grid search
from itertools import product

p = d = q = range(0, 2)
pdq = list(product(p, d, q))
```

```python
def train_test_split(series, train_size=.8):
    size = int(len(series)*train_size)
    train, test = series[:size], series[size:]

    print(f"Train set begins from {train.index[0].date()} to {train.index[-1].date()}, and is composed of {len(train)} observations.")
    print(f"Test set begins from {test.index[0].date()} to {test.index[-1].date()}, and is composed of {len(test)} observations.")
    return train, test
```

```python
train, test = train_test_split(amazon.close)
```

```python
# validation
train_fold, validation_fold = train_test_split(train)
```

```python
def predict_ARIMA(train, validation, arima_params):
    history = [x for x in train]
    predictions = list()
    progress = 1
    for i in range(len(validation)):
        # predict
        model = ARIMA(history, **arima_params)
        model_fit = model.fit(disp=0)
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        history.append(validation[i])
        print(f'one-step forward prediction, {progress} test samples out of {len(validation)}',end='\r')
        progress += 1
        
    return predictions
```

```python
params = {'order': (0, 1, 2)}
ypred = predict_ARIMA(train_fold, validation_fold, params)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1, 2, 1)
plt.title('Predictions Vs test')
plt.plot(validation_fold.index, validation_fold, label='test')
plt.plot(validation_fold.index, ypred, label='prediction')
plt.legend()

plt.subplot(1, 2, 2)
plt.title('Zoom in 2017 until 2018')
plt.plot(validation_fold.index, validation_fold, label='test')
plt.plot(validation_fold.index, ypred, label='prediction')
plt.xlim(pd.Timestamp('2017-05'), pd.Timestamp('2018-01'))
#plt.ylim((480, 600))
plt.legend()
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(validation_fold, ypred))
print('RMSE: %.3f' % rmse)
```

```python
# check residuals
# errors
residuals = [validation_fold[i]-ypred[i] for i in range(len(validation_fold))]
residuals=pd.Series(residuals)

#plot
fig, axes = plt.subplots(1, 2, figsize=(15,6))

fig = sm.graphics.tsa.plot_acf(residuals, lags=12, ax=axes[0])
fig = sm.graphics.tsa.plot_pacf(residuals, lags=12, ax=axes[1])
```

Residuals is white nose which means that the model has succeded to capture all underlying information /autcorrelation present in the timeseries.
